```python
import os
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
data_dir=os.getcwd()+'/data/'
```

```python
bob=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
```

```python
bob.head()
```

```python
countries_df=bob.copy()
```

```python
countries_df=bob.copy()
countries_df.drop(['Province/State','Lat','Long'], axis=1,inplace=True)
countries_df=countries_df.groupby('Country/Region').sum()
countries_df.sort_values('1/22/20',ascending=False,inplace=True)
countries_df.head()
```

```python
spread_df=countries_df.T
```

```python

plt.figure(figsize=(10,3.5),dpi=100)
plt.plot(spread_df.index,spread_df.China,color='r',label='China')
plt.plot(spread_df.index,spread_df.US,color='b',label='USA')
plt.plot(spread_df.index,spread_df.Italy,color='g',label='Italy')
plt.plot(spread_df.index,spread_df['Korea, South'],color='lightblue',label='South Korea')
plt.plot(spread_df.index,spread_df['Japan'],color='lightgrey',label='Japan')
plt.plot(spread_df.index,spread_df['Iran'],color='k',label='Iran')
plt.xticks(rotation=45)
plt.title('Corvid19 Confirmed Cases')
#plt.ylim(0,20000)
plt.legend(loc='best')
plt.show()
```

```python
spread_df
```

```python
def get_daily_new_cases(x):
    x_yesterday=x.shift(1)
    return x-x_yesterday

daily_cases=spread_df.copy()
for c in daily_cases.columns:
    daily_cases[c]=get_daily_new_cases(daily_cases[c])

daily_cases.to_csv(data_dir+'daily_cases_by_country.csv')
daily_cases.head()
```

```python
x_labels=[daily_cases.index[x] for x in range(len(daily_cases)) if x%2==0]
countries_to_plot=['US','Italy','Korea, South', 'United Kingdom']
colors=['navy','green','red','royalblue']
plt.figure(figsize=(10,3.5),dpi=100)

for i,country in enumerate(countries_to_plot):
    plt.plot(daily_cases.index,daily_cases[country],color=colors[i],lineWidth=2,label=f"{country}")

plt.plot(daily_cases.index,daily_cases['China'],color='red',lineStyle=':',lineWidth=1,label=f"{'China'}")
plt.xticks(x_labels,rotation=45)
plt.legend(loc='upper left',ncol=5)
plt.title('New Daily Reported Cases')
#plt.ylim(0,5000)
plt.show()
```

```python
china_provinces=bob.loc[bob['Country/Region']=='China',:].copy()
china_provinces.head()

china_provinces=china_provinces.drop(['Lat','Long','Country/Region'],axis=1).set_index('Province/State')
china_cum=china_provinces.T
china_cum.to_csv(data_dir+'china_provinces_cum.csv')
china_cum.head()
```

```python

big_5=china_cum.columns[:5]
xticks=[china_cum.index[x] for x in range(len(china_cum.index)) if x%2==0]

def make_simple_plot(df):
    for i,c in enumerate(big_5):
        plt.plot(df.index,df[c],label=c)
    
    plt.legend(loc='best')
    plt.xticks(xticks,rotation=45)
    return

plt.subplots(1,2,figsize=(10,3.5),dpi=100)

plt.subplot(1,2,1)
make_simple_plot(china_cum)

plt.subplot(1,2,2)
make_simple_plot(china_cum)
plt.ylim(0,2000)

plt.suptitle('Cummulative Cases')
plt.show()
```

## Make Daily File for China

```python
china_daily_cases=china_cum.copy()
for c in china_daily_cases.columns:
    china_daily_cases[c]=get_daily_new_cases(china_daily_cases[c])

china_daily_cases.to_csv(data_dir+'china_daily_cases.csv')
china_daily_cases.head()
```

```python
plt.subplots(1,2,figsize=(10,3.5),dpi=100)

plt.subplot(1,2,1)
make_simple_plot(china_daily_cases)

plt.subplot(1,2,2)
make_simple_plot(china_daily_cases)
plt.ylim(0,150)

plt.suptitle('Daily Cases China')
plt.show()
```
